#Importando bibliotecas

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

#Definindo e compilando a rede neural
- 1ª linha cria rede neural simples com uma unica camada
- 2ª linha compila a rede neural
  - Loss: mede as respostas adivinhadas em relação às respostas corretas e avalia o desempenho da qualidade
  - Optimizer: Com base na função de perda(loss), ele tenta minimizar a perda

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

#Fornecendo os dados

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

#Treinando a rede neural
- epochs: indica o número de vezes que o algoritmo de treinamento irá iterar sobre o conjunto de dados de treinamento completo durante o processo de treinamento da rede neural.

```python
model.fit(xs, ys, epochs=500)
```

#Usando o modelo
- Passando um numero para o modelo fazer uma previsão

```python
print(model.predict([10.0]))
```
